```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 1000)
y = x ** (-0.5) 

x_ = [1, 10, 50, 100, 300, 500, 800, 1000]
cpu = [0.386, 0.086, 0.066, 0.046, 0.013, 0.006, 0.003, 0.002]
cpu_u = [0.386, 0.011, 0.006, 0.026, 0.02, 0.02, 0.008, 0.013]
seq_rand = [0.386, 0.28,0.30,0.31,0.26,0.26,0.27,0.27]
seq_rand_seed = [0.386,0.28,0.14,0.25,0.26,0.27,0.25,0.26]
```

```python
plt.figure(figsize=(15,10))
plt.plot(x, y, label="origin")
plt.plot(x_, cpu, label="cpu")
plt.plot(x_, cpu_u, label="cpu_u")
plt.plot(x_, seq_rand, label="seq_rand") # rand()
plt.plot(x_, seq_rand_seed, label="seq_rand_seed") # gen.seed(static_cast<unsigned int>(std::time(0)))
plt.legend()
plt.xticks(np.arange(1, 1000, 50), np.arange(1, 1000, 50))
plt.ylim(0, 0.4)
```
